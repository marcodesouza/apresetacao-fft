# Manchas Solares


Os dados para este exemplo são do site <http://www.sidc.be/silso/datafiles> e compõem o arquivo sunspot_mediaAnual_1700-2015.dat.
Os dados foram modificados mantendo somente ocorrências do ano e da média demanchas solares para cada ano.

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
```

```python
ano, manchas = np.loadtxt('sunspot_mediaAnual_1700-2015.dat', delimiter=',', skiprows=0, unpack=True)
```

## Ciclo Solares de 1700 a 2015

```python
plt.plot(ano,manchas);
```

## Ciclo Solar 23 
Este ciclo é o anterior ao atual com duração de Maio 1996 a Janeiro 2008. O Ciclo atual (24) iniciou em 2008 com previsão de encerramento em 2020, atingindo seu auge em 2013.

```python
plt.plot(ano[297:308],manchas[297:308]);
```

```python
Y=np.fft.fftn(manchas)
n=Y.size
power=np.abs(Y[1:np.floor(n/2)])**2
nyquist=1./2
manchas = np.arange(1,n/2,1.)/(n/2)*nyquist
period = 1./manchas
```

```python
plt.plot(manchas,power) 
plt.ylabel('Quantidade');
plt.xlabel('Periodo (Ciclos/ano)');
```

```python
# pega o índice da potência mais alta
index=np.argmax(power)
```

```python
# nesta posição tem a frequência que ocorre o pico de potencia (?)
freq=manchas[index]
```

```python
# O ciclo ocorre como a inversa desta frequencia, ou seja aproximadamente 11 anos confirmando o Ciclo de Wolf
1/freq
```
